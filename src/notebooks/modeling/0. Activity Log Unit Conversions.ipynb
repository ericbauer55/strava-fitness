```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#pd.set_option("display.max_rows", 1000)
pd.set_option("display.max_columns", 100)
```

```python
file_paths = {'in':'../../../data/prepared/activity_log.csv',
              'out':'../../../data/prepared/activity_log_prepared.csv'
             }

df = pd.read_csv(file_paths['in'])

# guarantee the timestamps are datetime objects
time_columns = ['start_time','end_time','ride_date']

for time_col in time_columns:
    df[time_col] = pd.to_datetime(df[time_col])
```

```python
df.head()
```

# 1. Convert Strava Columns

```python
km2mi = 0.621371 # kilometers to miles
MpS2MPH = 2.23694 # meters per second to MPH
m2ft = 3.28084 # meters to feet

convert_dict = {'distance':km2mi,
                'max_speed':MpS2MPH,
                'avg_speed':MpS2MPH,
                'elevation_gain':m2ft,
                'elevation_loss':m2ft,
               }
```

```python
for column in convert_dict.keys():
    df[column] = df[column] * convert_dict[column]
```

```python
df.head()
```

# 2. Re-order the Columns

```python
df.columns
```

```python
col_order = ['ride_id','row_count','segment_count','ride_date','start_time','end_time','elapsed_time','ride_elapsed_time',
             'moving_time','ride_moving_time',
             'distance','elapsed_distance','elevation_gain','elapsed_ascent','elevation_loss','elapsed_descent',
             'elapsed_elevation', 'max_speed','ride_max_speed','avg_speed','ride_avg_speed','ride_cruise_speed',
             'avg_power','ride_avg_power','ride_max_power','peak_20min_power','training_window_id',
             'calories','perceived_exertion','simple_exertion','perceived_relative_effort'
            ]
```

```python
# check that I got everything
[x for x in col_order if x not in df.columns]
```

```python
df = df[col_order]
```
